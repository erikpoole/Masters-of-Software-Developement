```python
# imports and setup 

import pandas as pd
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt

import json
import itertools
import collections
```

```python
file = open("imdbClean.json", "r")
movies = []
for line in file:
    try:
        movies.append(json.loads(line))
    except:
        pass

print(len(movies))
```

```python
movies_USA = []
for movie in movies:
    try:
        if ('USA' in movie["Country"]):
            movies_USA.append(movie)
    except:
        pass

print(len(movies_USA))

# print(movies_USA[0])
```

```python
graph = nx.Graph()
appearance_counts = collections.Counter()
star_wars_actors = []

for movie in movies_USA:
    actors = movie["Actors"].split(', ')
    
    if movie["Title"] == 'Star Wars':
        star_wars_actors = actors
    
    for actor in actors:
        if actor != 'N/A':
            appearance_counts[actor] += 1
    for combination in itertools.combinations(actors, 2):
        graph.add_node(combination[0])
        graph.add_node(combination[1])
        graph.add_edge(combination[0], combination[1])
        

# print(len(appearance_counts))
# print(graph.nodes)
# print()
# print(graph.edges)
```

```python
print("Most Common American Actor:")
print(appearance_counts.most_common(1))
```

```python
distant_bacon_actors = {}
close_bacon_actors = {}
for actor in appearance_counts.keys():
    try:
        path = nx.shortest_path(graph, source='Kevin Bacon', target=actor, weight=None, method='dijkstra')
        if (len(path) < 7):
            close_bacon_actors[actor] = path
        else:
            distant_bacon_actors[actor] = path
    except:
        distant_bacon_actors[actor] = "No Path"

print(len(appearance_counts))
print(len(distant_bacon_actors))
print("Actors outside of Kevin Bacon's Sphere of Light and Goodness:")
for actor in distant_bacon_actors:
    print(actor, ": ", distant_bacon_actors[actor])
```

```python
actors_sw_paths = {}
for sw_actor in star_wars_actors:
    for actor in appearance_counts.keys():
        try:
            path = nx.shortest_path(graph, source=sw_actor, target=actor, weight=None, method='dijkstra')
            if actor not in actors_sw_paths:
                actors_sw_paths[actor] = path
            else:
                if actors_sw_paths[actor] == "No Path" or len(actors_sw_paths[actor]) > path :
                    actors_sw_paths[actor] = path

        except:
            if actor not in actors_sw_paths:
                actors_sw_paths[actor] = "No Path"

print(len(appearance_counts))
count = 0
for actor in actors_sw_paths:
    if actors_sw_paths[actor] == "No Path" or len(actors_sw_paths[actor]) > 6:
        count += 1     
print(count)

print("Actors outside of the Death Star's Range:")
for actor in actors_sw_paths:
    if actors_sw_paths[actor] == "No Path" or len(actors_sw_paths[actor]) > 6:
        print(actor, ": ", actors_sw_paths[actor])
```

```python
bacon_graph = nx.ego_graph(graph, 'Kevin Bacon', radius=1, center=True, undirected=False, distance=None)

plt.figure(figsize=(15,15))
pos = nx.circular_layout(bacon_graph, scale=1.1)
nx.draw_networkx_labels(bacon_graph, pos)
nx.draw_circular(bacon_graph, node_size=1)

plt.show()
```

```python
actor_pageRanks = nx.pagerank(graph)

# Arbitrary starting point
maxActor = 'Kevin Bacon'
for actor in actor_pageRanks:
    if actor_pageRanks[actor] > actor_pageRanks[maxActor]:
        maxActor = actor

print(maxActor, actor_pageRanks[maxActor])
```

```python
movie_based_graph = nx.Graph()
appearance_counts = collections.Counter()

actors_to_movies = {}
for movie in movies_USA:
    movie_based_graph.add_node(movie['Title'])
    
    actors = movie["Actors"].split(', ')
    for actor in actors:
        if actor != 'N/A':
            if actor not in actors_to_movies:
                actors_to_movies[actor] = [movie['Title']]
            else:
                working = actors_to_movies[actor]
                working.append(movie['Title'])
                actors_to_movies[actor] = working

# print(actors_to_movies['Kevin Bacon'])
# print(movie_based_graph.nodes)

for actor in actors_to_movies:
    for combination in itertools.combinations(actors_to_movies[actor], 2):
        movie_based_graph.add_edge(combination[0], combination[1])
    
# print(movie_based_graph.edges('The Steel Helmet'))
```

```python
movie_pageRanks = nx.pagerank(movie_based_graph)

# Arbitrary starting point
maxMovie = 'Star Wars'
for movie in movie_pageRanks:
    if movie_pageRanks[movie] > movie_pageRanks[maxMovie]:
        maxMovie = movie

print(maxMovie, movie_pageRanks[maxMovie])
```

### What does the page rank tell you here? ###

**Page rank is related to the degree of the node that we calculated before, but is more complicated, because it cares about the degree of its neighbors as well.  The more central and connected a node the better its pagerank.  For Oliver Hardy, that means he was probaly in a lot of movies that had actors go on to do make a lot more movies.  For Space Cowboys that probably means that it had a lot of actors that were in other movies with famous actors**

```python
count_distant = 0
sum_distant = 0
for actor in distant_bacon_actors:
    if distant_bacon_actors[actor] == "No Path":
        count_distant += 1
        try:
            sum_distant += graph.degree(actor);
        except:
            pass

average_distant = sum_distant / count_distant
```

```python
count_close = 0
sum_close = 0
for actor in close_bacon_actors:
    count_close += 1
    sum_close += graph.degree(actor);

average_close = sum_close / count_close
```

```python
print(average_distant)
print(average_close)
```
